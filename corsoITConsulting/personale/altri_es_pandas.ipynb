**Pandas Data Frame**

```python
# Create a dictionary with some different data types as values

my_dict = {"name" : ["Joe","Bob","Frans"],
           "age" : np.array([10,15,20]),
           "weight" : (75,123,239),
           "height" : pd.Series([4.5, 5, 6.1], 
                                index=["Joe","Bob","Frans"]),
           "hair_color" : pd.Series(["brown","black","blonde"], 
                                index=["Joe","Bob","Frans"]),
           "siblings" : 1,
           "gender" : "M"}

df = pd.DataFrame(my_dict)   # Convert the dict to DataFrame

df                           # Show the DataFrame
```

```python
#call a column as in a dictionary
print(df["name"])
#or as an attribute
df.name
```

```python
del df['hair_color']  #delete a column√π
df
```

```python
#add column
df['new_column'] = pd.Series([1,2], index=["Joe","Bob"])
df
```
