```python
import numpy as np
import pandas as pd
```

```python
titanic_train = pd.read_csv("titanic.csv")      # Read the data

char_cabin = titanic_train["Cabin"].astype(str)    # Convert cabin to str

new_Cabin = np.array([cabin[0] for cabin in char_cabin]) # Take first letter

titanic_train["Cabin"] = pd.Categorical(new_Cabin)  # Save the new cabin var
```

```python
titanic_train.dropna(inplace=True)  # Drop NA values
```

```python
titanic_train.columns
```

**Cross Tables (tabelle di contingenza)**

```python
my_tab = pd.crosstab(index = titanic_train["Survived"], columns = "counts")  # Crosstab
my_tab
```

```python
my_tab = pd.crosstab(index = titanic_train["Cabin"], columns = "counts")  # Crosstab
my_tab
```

```python
my_tab = pd.crosstab(index = titanic_train["Sex"], columns = "counts")  # Crosstab
my_tab
```

```python
my_tab_perc = my_tab/my_tab.sum()  # Crosstab
my_tab_perc
```

```python
  # Create age categories
titanic_train["Age Catergory"] = pd.cut(titanic_train["Age"], bins=[0,12,20,40,60,80], labels=["Child","Teenager","Adult","Middle_Aged","senior"])  # Create age categories
titanic_train["Age Catergory"]
```

```python
titanic_train["Age Category"] = [
    "Child"        if age <= 12 else
    "Teenager"     if age <= 20 else
    "Adult"        if age <= 40 else
    "Middle_Aged"  if age <= 60 else
    "Senior"       
    for age in titanic_train["Age"]
]
```

```python
survived_class = pd.crosstab(index = titanic_train["Survived"], columns = titanic_train["Pclass"], margins = True)  # Crosstab
survived_class.index = ["Not Survived", "Survived", "Total row"]
survived_class.columns = ["class 1", "class 2", "class 3", "Total Col"]
survived_class
```

```python
survived_class_perc = survived_class/ survived_class.loc["Total row","Total Col"]  # Percentage table
survived_class_perc
```

```python
my_tab = pd.crosstab(index = titanic_train["Age Catergory"], columns = "counts")  # Crosstab
my_tab
```

```python
my_tab = pd.pivot_table(
    titanic_train,
    index="Age Catergory",
    columns="Pclass",
    values="PassengerId",   # qualsiasi colonna numerica va bene
    aggfunc="count",
    margins=True
)

my_tab
```
